# ASSIGNMENT OF SESSION 4 (LINE & BAR PLOT)


```python
# Importing the seaborn & Matplotlib libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Creating a count plot for finding the academic status of students

status = sns.countplot(x="Qualification_completed", data=data_of_chilla, order=["Pre-Bachelors","Bachelors","Masters","PhD"])
plt.title("Academic Status of Students")
status
```

- Result : This graph shows that maximum number or students are those who have completed their Masters.

```python
# Importing the seaborn & Matplotlib libraries
import seaborn as sns
import matplotlib.pyplot as plt

# CHanging Figure size
plt.figure(figsize=(8,6))

# Creating a count plot for finding the academic status of students

status = sns.barplot(x="How many hours you code a day? (int) e.g: 5,4,3",y="Qualification_completed", data=data_of_chilla, order=["Pre-Bachelors","Bachelors","Masters","PhD"], ci =None )
plt.title("Field of Study of Students")
status
```
