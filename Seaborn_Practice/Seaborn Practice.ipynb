# Introduction to Seaborn
   
   Seaborn is a library for making statistical graphics in Python. It builds on top of matplotlib and integrates closely with  pandas data structures.Seaborn helps you explore and understand your data.
   
# 4 Basic Steps for Seaborn Plotting

- Step # 1 : Import the libraries (Like : import seaborn as sns)
- Step # 2 : Set the theme i.e. In our classes we told about the canvas (Like : sns.set_theme())
- Step # 3 : Load the data set (Like : sns.load_dataset("data_which_you_want_to_load"))
- Step # 4 : Create Plot (Like : sns.relplot(x=" ",y=" ",data=data_which_you_want_to_load)

```python
# Importing Libraries

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Setting up the theme

sns.set_theme(style = "dark")

# Loading a data set

tips = sns.load_dataset("tips")
tips
```

# Scatter Plot 

This is a simple way to plot a scatter plot. In this we only give the x & y axis and just tell the function that data set which you should have to pick is tips.

```python
sns.scatterplot(
    x="total_bill",
    y="tip",
    data = tips)
```

# Adding "hue" to the above Plot

```python
sns.scatterplot(
    x="total_bill",
    y="tip",
    hue="time",
    data = tips)
```

# Adding "style" to the Plot
 -  By adding hue command the plot adds some colors making the graph more easy to analyze but just focus on the pointer/marker of the data points they all are circle, we can make it more attract ful by changing the style of the marker using style command.
 - By giving style = "time" it will change the shape of the marker. 

```python
sns.scatterplot(
    x="total_bill",
    y="tip",
    hue="time",
    style="time",
    data = tips)
```

# Assigning "hue" & "style" with different variables
- Assigning hue and style to different variables will vary colors and markers independently

```python
sns.scatterplot(
    x="total_bill",
    y="tip",
    hue="day",
    style="time",
    data = tips)
```

In above plots we assigned a catagorical data to hue but if we assigned a numeric data to hue then the semantic mapping will be quantitative and use a different default palette

```python
sns.scatterplot(
    x="total_bill",
    y="tip",
    hue="size",
    data = tips)
```

# Including "palette" command

- Palette force categorical mapping of the hue variable adding different colors in the plot.

```python
sns.scatterplot(
    x="total_bill",
    y="tip",
    hue="size",
    palette="deep",
    data = tips)
```

# Adding "size" command to the plot
- Size command makes the marker size different to make the plot more readable

```python
sns.scatterplot(
    x="total_bill",
    y="tip",
    hue="size",
    size="size",
    data = tips)
```
