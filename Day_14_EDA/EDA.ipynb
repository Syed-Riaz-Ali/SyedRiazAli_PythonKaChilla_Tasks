# Cleaning and Filtering the data

> Step-1 : Finding the missing vlaues

```python
kashti.isnull()
```

```python
# Finding out the how many null values are there

kashti.isnull().sum()
```

```python
# dropping deck column because it has many null values

kashti_new = kashti.drop(["deck"], axis=1)
kashti_new
```

```python
# Now in the coulmn age we have many NaN values we need to drop those cells as well so we use the following command for that

ks = kashti_new.dropna()
ks
```

```python
ks.shape
```

```python
# To find out that how many males & females are there in our data set so we use .value_counts() function

ks["sex"].value_counts() 
```

#### Finding out the outlier in the dataset

```python
ks.columns
```

```python
# To findout the outliers we use the help of boxplot

sns.set_style("darkgrid")

sns.boxplot(data = ks, x = "sex", y = "age")
```

```python
sns.boxplot(data = ks, y = "age")
```

```python
# Making distplot to elaborate the outliers more further

sns.distplot(ks["age"])
```

#### Removing the outliers from the data set 

```python
ks["age"].mean()
```

```python
ks_otliers = ks["age"] < 68
ks_otliers
```

```python
ks = ks.loc[ks_otliers]
ks
```

```python
sns.boxplot(data = ks, y = "age")
```

```python
ks["age"].mean()
```

```python
# Creating box plot of all the coulmns in the data set having numerical values

ks.boxplot()
```

>The above graph shows the outliers of differrent columns

```python
# Dropping the rows which are outliers in fare column

ks = ks[ks["fare"] < 300]
ks
```

# Checking the relationship of dataset 

```python
# Making a corelation matrix

cor_ks_clean = ks.corr()
cor_ks_clean
```

```python
sns.heatmap(cor_ks_clean)
```

```python
# To add the value of corelation

plt.figure(figsize=(15,10))

sns.heatmap(cor_ks_clean, annot = True)
```
