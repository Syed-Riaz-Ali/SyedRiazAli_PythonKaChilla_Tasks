## ***Data Normalization***


>
> Our data in column ***age*** & ***fare*** are dispersed so we have to normalize it by using the below methods of Normalization
> 

### ***Methods of Normalization***

#### ***1. Simple feature Scaling***

$$x(new) = \frac{x(old)}{x(max)}$$

```python
ksh_sfs = ksh1
ksh_sfs.head() 
```

```python
ksh_sfs["fare"] = ksh_sfs["fare"]/ksh_sfs["fare"].max()
ksh_sfs["age"] = ksh_sfs["age"]/ksh_sfs["age"].max()
ksh_sfs.head()
```

```python
sns.set_style("darkgrid")

ksh_sfs["age"].hist()
```

```python
ksh_sfs["fare"].hist()
```

#### ***2. Min-Max Method*** 

$$x(new) = \frac{x(old) - x(min)}{x(max)- x(min)}$$

```python
ksh_min_max = ksh1
ksh_min_max["fare"] = (ksh_min_max["fare"] - ksh_min_max["fare"].min())/(ksh_min_max["fare"].max() - ksh_min_max["fare"].min())
ksh_min_max["age"] = (ksh_min_max["age"] - ksh_min_max["age"].min())/(ksh_min_max["age"].max() - ksh_min_max["age"].min())
ksh_min_max.head()
```

```python
ksh_min_max["age"].hist()
```

```python
ksh_min_max["fare"].hist()
```

#### ***3. Z-Score Method***

$$x(new) = \frac{x(old) - x(mean)}{x(std)}$$

```python
ksh_z_score = ksh1

ksh_z_score["fare"] = (ksh_z_score["fare"]-ksh_z_score["fare"].mean())/ksh_z_score["fare"].std()
ksh_z_score["age"] = (ksh_z_score["age"]-ksh_z_score["age"].mean())/ksh_z_score["age"].std()
ksh_z_score.head()
```

```python
ksh_z_score["age"].hist()
```

```python
ksh_z_score["fare"].hist()
```

```python
plt = sns.histplot(data = ksh_z_score, x= "age")
plt
```

```python
plt = sns.histplot(data = ksh_z_score, x= "fare")
plt
```

#### ***4. Log  Transformation Method***

$$x(new) = log (x(old))$$

```python
ksh_log = ksh2

ksh_log["fare"] = np.log(ksh_log["fare"])
ksh_log["age"] = np.log(ksh_log["age"])
ksh_log.head()
```

```python
ksh_log.isnull().sum()
```

```python
ksh_log = ksh_log.dropna()
```

```python
ksh_log.shape
```

```python
plt = sns.histplot(data = ksh_log, x= "age")
plt
```

```python
plt = sns.histplot(data = ksh_log, x= "fare")
plt
```
