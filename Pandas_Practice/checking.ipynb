```python
# Importing libraries

import pandas as pd
import numpy as np

# Loading the data base of population

p_data = pd.read_csv("FAOSTAT_data_1-12-2022.csv")
p_data
```

```python
pak = pd.DataFrame(p_data, columns=["Area Code (FAO)","Area","Year","value"])
pak
```

```python
df = pak.set_index("Area Code (FAO)")
df
```

```python
keys = {"Afghanistan": 2, "USSR" : 228, "Austrailia" : 10, "Bangladesh" : 16, "Brazil" : 21, "Canada" : 33, "China" : 351, "Finland" : 67, "France" : 68,
        "Germany" : 79, "Greece" : 84, "India" : 100, "Indonesia" : 101, "Iran (Islamic Republic of)" : 102, "Israel" : 105, "Italy" : 106, "Japan" : 110,
        "Mexico" : 138, "New Zealand" : 156, "Pakistan" : 165, "Portugal" : 174, "Russian Federation": 185, "Saudi Arabia" : 194, "Singapore" : 200, 
        "South Africa" : 202, "Srilanka" : 38, "Turkey" : 223, "United Arab Emirates" : 225, "United Kingdom of Great Britain and Northern Ireland": 229,
        "United States of America": 231}
keys
```

```python
yiu = pd.Series(keys)
yiu
```

```python
w = yiu["Turkey"]
w
```

```python
df_Afgh = df.loc[w]
df_Afgh
```

```python
df_year = df_Afgh.set_index("Year")
df_year
```

```python
year_1950 = df_year.loc[1950]
year_1950
```

```python
y = np.arange(1950, 2019, step =1)
y
```

```python
ay = np.delete(y, 68)
ay
```

```python
year_pop_Afgh = []
for i in range(68):
    y1 = df_year.loc[y[i]]
    pop = y1["value"].sum()
    year_pop_Afgh.append(pop.round())

year_pop_Afgh
```

```python
b = np.array(year_pop_Afgh)
b
```

```python
o = pd.DataFrame({"Year" : ay, "Population": b})
o
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("darkgrid")

pop_plot_of_afgh = sns.lineplot(data=o, x = "Year", y="Population")
plt.title("Population of Afghanistan")
pop_plot_of_afgh
```
